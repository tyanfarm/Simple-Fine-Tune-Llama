```python
import json

file_path = "./hotel_addresses_updated.json"
hotel_names = []


with open(file_path, 'r', encoding='utf-8') as file:
    raw_data = json.load(file)

for data in raw_data:
    print(data['name'])
    hotel_names.append(data['name'])
```

```python
file_addresses_path = "./finetuned-datasets/5-hotel_addresses.json"

with open(file_addresses_path, 'r', encoding='utf-8') as file:
    addresses_data = json.load(file)

for data in addresses_data:
    print(data)
```

```python
import math 

need = math.ceil(len(addresses_data) / 10)

if len(hotel_names) < need:
    raise ValueError(f"Cần ít nhất {need} tên trong B, hiện chỉ có {len(hotel_names)}")

for i, item in enumerate(addresses_data):
    item["hotel_name"] = hotel_names[i // 10]

print(addresses_data)
```

```python
from typing import List, Dict

def save_to_file(data: List[Dict], filename: str):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write("[")

        for index, entry in enumerate(data):
            char = ','
            if index == len(data) - 1:
                char = ''

            file.write(json.dumps(entry, ensure_ascii=False) + f"{char}\n")

        file.write("]")
        
    print(f'Saved {len(data)} entries to {filename}')
```

```python
save_to_file(addresses_data, '5-hotel.json')
```
